```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(Z):
    e_z = np.exp(Z)
    A = e_z / e_z.sum(axis=1, keepdims=True)
    return A
def softmax_temp(Z, T):
    e_z = np.exp(Z/T)
    A = e_z / e_z.sum(axis=1, keepdims=True)
    return A
```

```python
X = np.random.randn(1, 10)
```

```python
T = [0.5, 1, 2, 10]

plt.figure(figsize=(12,4))
for t in T:
    # plt.plot(softmax(X).reshape(-1), label='T=1')
    plt.plot(softmax_temp(X, t).reshape(-1), label='T={}'.format(t))
# plt.ylim([0,1])
plt.legend(loc='best')
plt.title('Softmax distribution score')
plt.savefig('Softmax distribution score',dpi=400)
plt.show()
```
